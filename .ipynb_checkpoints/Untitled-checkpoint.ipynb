```python
session= tf.Session()
```

```python
session.run(b)
```

## now we need to close the session to release the used resources

```python
session.close()
```

# we need to close every time which is not convenient

```python
tf.device("cpu0")
with tf.Session() as session:
    print(session.run(b))
    
tf.device("CPU 1")
with tf.Session() as session:
    print(session.run(tf.subtract(tf.constant(14),tf.constant(14))))
```

# variable and constant

```python
state= tf.Variable(2)
cont=tf.constant(5)
new_value=tf.add(state,cont)
state=tf.assign(state,new_value)
```

```python
in_state=tf.global_variables_initializer()
```

```python
with tf.Session() as session:
     session.run(in_state)
     print(session.run(state))
```

## placeholder is use to take value from outside the model

```python
a=tf.placeholder(tf.complex128)
```

```python
b=2*a
```

```python
print(b)
```

### since we have make a hole in our model using placeholder we need to pass additional value from outside

```python
with tf.Session() as session:
    print(session.run(b,feed_dict={a:3.5}))
```

```python
dictionary={a:[[1,2,3],[5,4,3]]}
with tf.Session() as sess:
        print(sess.run(a,feed_dict=dictionary))
```

## now take an example


```python
x= tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
```

```python
import numpy as np
```

```python
rand_x= np.random.uniform(10,100,(5,5))
print(rand_x)
print('\n')
rand_y= np.random.uniform(100,200,(5,5))
print(rand_y)
```

```python
sum = x+y
```

```python
with tf.Session() as sess:
    print(sess.run(sum,feed_dict={x:rand_x,y:rand_y}))
```
