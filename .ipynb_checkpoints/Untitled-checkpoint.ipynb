# variable and constant

```python
state= tf.Variable(2)
cont=tf.constant(5)
new_value=tf.add(state,cont)
state=tf.assign(state,new_value)
```

```python
in_state=tf.global_variables_initializer()
```

```python
with tf.Session() as session:
     session.run(in_state)
     print(session.run(state))
```

## placeholder is use to take value from outside the model

```python
a=tf.placeholder(tf.complex128)
```

```python
b=2*a
```

```python
print(b)
```

### since we have make a hole in our model using placeholder we need to pass additional value from outside

```python
with tf.Session() as session:
    print(session.run(b,feed_dict={a:3.5}))
```

```python
dictionary={a:[[1,2,3],[5,4,3]]}
with tf.Session() as sess:
        print(sess.run(a,feed_dict=dictionary))
```

# Basic Regression


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x_data=np.linspace(1,11,10) + np.random.uniform(-0.4,0.5,10)
x_data
```

```python
y_data=np.linspace(1,11,10)+ np.random.uniform(-0.4,0.5,10)
y_data
```

```python
plt.plot(x_data,y_data,"*")
```

```python
np.random.rand(2)
```

```python
# y=mx+b
m= tf.Variable(0.89)
b=tf.Variable(0.35)
```

```python
init= tf.global_variables_initializer()
```

```python
for x,y in zip(x_data,y_data):
    error=0
    error+=(y-(m*x+b))**2
```

```python
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.001)
train=optimizer.minimize(error)
```

```python
with tf.Session() as sess:
    sess.run(init)
    print(sess.run(error))
    for i in range(40):
        sess.run(train)
    fslope,Finter=sess.run([m,b])
```

```python
fslope
```

```python
Finter
```

```python
x_test=np.linspace(-1,11,10)
y_predict=fslope*x_test+Finter

plt.plot(x_test,y_predict,'r')
plt.plot(x_data,y_data,'*')
```
